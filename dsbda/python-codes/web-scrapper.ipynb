```python
# Questions-> 
# create a review scrapper for any ecommerce website to fetch real time comments, reviews, and customer name using python. 
# create a review scrapper for any ecommerce website to fetch real time ratings, comment tags, and customer name using python. 
import requests
import bs4
```

```python
request1 = requests.get("https://www.flipkart.com/poco-c31-royal-blue-64-gb/p/itm19effae969b86?pid=MOBG73E7GKQK4KZP&lid=LSTMOBG73E7GKQK4KZPR5ICMK&marketplace=FLIPKART&q=poco+c31&store=tyy%2F4io&srno=s_1_4&otracker=search&otracker1=search&fm=organic&iid=f64462e5-f434-496e-9eac-b686c765a3ce.MOBG73E7GKQK4KZP.SEARCH&ppt=pp&ppn=pp&ssid=pazolmu31c0000001685195889696&qH=8a1462a8a580f1cb")
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.find_all("div",{'class':"t-ZTKy"}) 

for review in reviews:
  print(review.get_text() + "\n\n")
```

```python
# Fetching rating
ratings = soup.find("div",{"class":"_2d4LTz"}).get_text()
print(ratings)
```

```python
# Individual ratings

individual_ratings = soup.find_all("div",{"class":"_3LWZlK _1BLPMq"})

for rating in individual_ratings:
  print(rating.get_text()+"\n")
```

```python
# Comment Tags

tags = soup.find("span",{"class":"yhB1nd GXgmTe"})

print(tags.get_text())
```

```python
# Fetching customer name

names = soup.find_all("p",{"class":"_2sc7ZR _2V5EHH"})

for name in names:
  print(name.get_text())
```
