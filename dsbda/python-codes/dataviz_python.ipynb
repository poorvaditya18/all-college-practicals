# Data Visualization

```python
'''
Question -> Visualise the data usin python libraries matplotlib , seaborn by plotting the graphs for FB metrics/ Air Quality / Heart disease/ Iris dataset.
            - Draw box plot 
            - histogram 
            - single line graph 
            - multiline graph .
'''

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/dataset_Facebook.csv",sep=";")
df
```

## Boxplot in Seaborn

```python
sns.boxplot(data=df, x="Type", y="comment")
plt.ylim(0,30)
plt.show()
```

```python
sns.boxplot(data=df, x="Type", y="Total Interactions")
plt.ylim(0,1000)
plt.figaspect(1.8)
plt.show()
```

## Boxplot in Matplotlib

```python
plt.boxplot(x=[df["Post Month"],df["Post Weekday"]],vert=True)
plt.show()
```

## Histogram in Seaborn

```python
sns.histplot(data=df, x="Type",kde=True)
```

```python
sns.histplot(data=df, x="Post Month", kde=True, hue="Type")
```

## Histogram in Matplotlib

```python
plt.hist(data=df, x="Type")
plt.show()
```

## Line Graphs in Seaborn

### Single line graph

```python
sns.lineplot(data=df, x="comment", y="Total Interactions")
```

### Multiline Graph

```python
columns = ["like","share","comment"]
sns.lineplot(data=df[columns])
```

## Line Graphs in Matplotlib


### Single line graph

```python
sns.heatmap(df.corr(), annot=True, cmap="cool")
```

```python
plt.plot(df["Page total likes"],df["Post Month"])
```

### Multiline Graph

```python
plt.plot(df["Post Month"], df["share"],label="Shares")
plt.plot(df["Post Month"],df["comment"],label="Comment")
plt.plot(df["Post Month"],df["like"],label="Likes")
plt.legend()
# plt.ylim(0,1000)
plt.show()
```
